```python
# khởi tạo RDD từ Spark Driver
data = sc.parallelize([1,2,3,4,5,5])
data.collect()
```

```python
# khởi tạo RDD từ một RDD khác
# map để chuyển sang một RDD khác theo cặp key - value
# reduceByKey để tính tổng RDD 
# sortBykey sắp xếp các giá trị với True là từ bé đến lớn
data1 = data.map(lambda x: (x,1)).reduceByKey(lambda x,y: x+y).sortByKey(True)
data1.collect()
```

```python
data2 = data.map(lambda x: x*2)
data2.collect()
```

```python
# union gom 2 RDD lại với nhau
data3 = data.union(data2)
data3.collect()
```

```python
# sample lấy tập con ngẫu nhiên từ tập dataset lớn
# gồm 2 argument : 
# withReplacement: mặc định là false, True dùng để repeat value
# fraction : range [0.0, 1.0] ~ 0% - 100%.
# Every time you run a sample() function it returns a different set of sampling records
data4 = data.sample(True, 0.7)
data4.collect()
```

```python
# mapPartion dùng giống map nhưng tác động trên từng cụm
# Filter để lọc điều kiện 
dataFilter = data3.filter(lambda x : x%2 == 0)
dataFilter.collect()
# dataFilter.count() # đếm
# dataFilter.take(5) # trả về số giá trị muốn trả về
```

```python
# join lấy các key trùng nhau
data5 = sc.parallelize([(1,"a"),(2,"b"),(3,"c")])
data6 = sc.parallelize([(1,"d"),(2,"e"),(3,"c")])
data7 = data5.join(data6)
data7.collect()
```

```python
# intersection chỉ ra phần tử giống nhau giữa 2 dataset
data8 = data5.intersection(data6)
data8.collect()
```

```python
# accumulator used shared variables to perform sum and counter operations similar to Map-reduce 
# c1 use lambda
accum = sc.accumulator(0)
rdd = sc.parallelize([1,2,3,4,5])
rdd.foreach(lambda x: accum.add(x))
print(accum.value)
```

```python
# c2 use for loop
accuSum=sc.accumulator(0)
def countFun(x):
    global accuSum
    accuSum+=x
rdd.foreach(countFun)
print(accuSum.value)
```

```python
# using accumCount: get value index * with sum element
accumCount=sc.accumulator(0)
rdd2=sc.parallelize([1,2,3,4,5])
rdd2.foreach(lambda x:accumCount.add(3))
print(accumCount.value)
```
